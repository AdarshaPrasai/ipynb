```python
ms = pd.read_csv('medical_students.csv')
```

```python
ms.head()
```

```python
ms.isnull().sum()/len(ms)*100
```

```python
ms.info()
```

```python
numeric = ms.select_dtypes(exclude='object')
categoric = ms.select_dtypes(include='object')
```

```python
numeric.fillna(numeric.mean(),inplace=True)
```

```python
numeric.isnull().sum()
```

```python
categoric.apply(lambda col:col.fillna(col.mode()[0],inplace=True))
```

```python
categoric.isnull().sum()
```

```python
combine_data = pd.concat([numeric,categoric],axis=1)
```

```python
combine_data.head()
```

```python
combine_data.duplicated().sum()
```

```python
combine_data.drop_duplicates(inplace=True)
```

```python
categoric.columns
```

```python
combine_data.duplicated().sum()
```

```python
encoded_data = pd.get_dummies(combine_data,columns=['Gender', 'Blood Type', 'Diabetes', 'Smoking'],drop_first=True,dtype='int')
```
